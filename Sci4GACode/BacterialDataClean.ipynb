```python
import pandas as pd
import numpy as np
```

```python
bacterialData = pd.read_csv("/Users/keerthyrangan/Downloads/bacteria(Sheet1).csv", encoding = "latin-1")
bacterialData
```

```python
variables_to_keep = ["EventDate","GroupName","SiteName","SiteID","site_latitude","site_longitude","site_altitude","colony_avg"]
bacterialData = bacterialData[variables_to_keep]
bacterialNew = bacterialData.dropna()

bacterialNew
```

```python
bacterialNew["colony_avg"] = pd.to_numeric(bacterialNew["colony_avg"], errors="coerce")
bacterialNew = bacterialNew.dropna()
bacterialNew
```

```python
from sklearn.ensemble import IsolationForest

colony_data = np.array(bacterialNew["colony_avg"]).reshape(-1,1)
clf = IsolationForest(contamination=0.1)
outliers = clf.fit_predict(colony_data)
outliers =  colony_data[outliers == -1].flatten()

for i in bacterialNew.index:
    # Check if the value is an outlier
    if bacterialNew["colony_avg"][i] in outliers:
        # Drop the row with the outlier
        bacterialNew = bacterialNew.drop(i)

bacterialNew
```

```python
bacterialNew["EventDate"] = pd.to_datetime(bacterialNew["EventDate"])

colonies2022Data = bacterialNew[bacterialNew["EventDate"].dt.year == 2022]
colonies2022Data
```

```python
colonies2022Data.to_csv('/Users/keerthyrangan/Downloads/colonies2022Data.csv', index=False)
```
