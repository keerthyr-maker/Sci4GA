```python
import pandas as pd
import numpy as np
```

```python
chemicalsData = pd.read_csv("/Users/keerthyrangan/Downloads/chemicalData.csv")
chemicalsData
```

```python
variables_to_keep = ["EventDate","GroupName","SiteName","SiteID","site_latitude","site_longitude","site_altitude","air_temp","water_temp"]
tempData = chemicalsData[variables_to_keep]
```

```python
tempNew = tempData.dropna()
```

```python
from sklearn.ensemble import IsolationForest

water_data = np.array(tempNew["water_temp"]).reshape(-1,1)
clf = IsolationForest(contamination=0.1)
outliers = clf.fit_predict(water_data)
outliers =  water_data[outliers == -1].flatten()

for i in range(len(tempNew["water_temp"])):
    if i in tempNew.index and tempNew["water_temp"][i] in outliers:
        tempNew["water_temp"].drop(i)

tempNew
```

```python
tempNew["EventDate"] = pd.to_datetime(tempNew["EventDate"])

temp2022Data = tempNew[tempNew["EventDate"].dt.year == 2022]
temp2022Data
```
